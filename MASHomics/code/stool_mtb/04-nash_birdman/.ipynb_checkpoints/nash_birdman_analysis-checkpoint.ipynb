# Birdman IHC plotting

```python
import pandas as pd
import qiime2 as q2
import numpy as np
import biom

# plotting 
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def read_results(p):
    return pd.read_csv(p, sep='\t', index_col='Feature')
def unpack_hdi_and_filter(df, col):
    df[['lower', 'upper']] = df[col].str.split(',', expand=True)
    # remove ( from lower and ) from upper and convert to float 
    df.lower = df.lower.str[1:].astype('float')
    df.upper = df.upper.str[:-1].astype('float')
    
    df['credible'] = np.where((df.lower > 0) | (df.upper < 0), 'yes', 'no')
    
    df.upper = df.upper - df[col.replace('hdi', 'mean')]
    df.lower = df[col.replace('hdi', 'mean')] - df.lower
    
    return df
```

## Metabolomics

```python
ihc=pd.read_csv('/home/sfloresr/projects/IHC-Amulya/birdman_results/inferences/inferences-results/quantification_table-rmft-clean-airihc.beta_var.tsv', sep='\t', index_col='Feature')
ihc
```

```python
data_dict = {
    'Air_IHC': '/home/sfloresr/projects/IHC-Amulya/birdman_results/inferences/inferences-results/quantification_table-rmft-clean-airihc.beta_var.tsv', 
    'Air_BSH': '/home/sfloresr/projects/IHC-Amulya/birdman_results/inferences/inferences-results/quantification_table-rmft-clean-airBSH.beta_var.tsv',
    'IHC_BSH': '/home/sfloresr/projects/IHC-Amulya/birdman_results/inferences/inferences-results/quantification_table-rmft-clean-IHCBSH.beta_var.tsv'
}

vars_to_check = {
    'Air_IHC' : 'C(combined_condition, Treatment(\'Air_BSH_neg\'))[T.IHC_BSH_neg]_',
    'Air_BSH' : 'C(combined_condition, Treatment(\'Air_BSH_neg\'))[T.Air_BSH_pos]_',
    'IHC_BSH' : 'C(combined_condition, Treatment(\'IHC_BSH_neg\'))[T.IHC_BSH_pos]_'
}
```

```python
# update names 
for k in data_dict.keys():
    df = read_results(data_dict[k])
    for v in vars_to_check: 
        if v in k: 
            var = vars_to_check[v] 
    print(k)
    print('Unfiltered Shape:  ' + str(df.shape))
    sub_df = unpack_hdi_and_filter(df, var+'hdi')
    print('Filtered Shape: ' + str(sub_df.loc[sub_df['credible'] == 'yes'].shape))
    sub_df.rename_axis(index='Feature', inplace=True)
    
    data_dict[k] = sub_df.sort_values(by=var+'mean')
```

```python
data_dict
```

```python
# check if there are more than 20 features --> if so, only display top and bottom 10 
# otherwise display all 

def display_top_n_feats(df, n, yvar, xvar, xlab, ylab, title):
    if df.shape[0] < 2*n: 
        df_for_display = df 
    else: 
        bottomn = df[:n]
        topn = df[-1*n:]
        df_for_display = pd.concat([bottomn, topn])
    df_for_display['Feature']=df_for_display['Feature'].apply(str)
    sns.stripplot(data=df_for_display, y=yvar, x=xvar)
    plt.errorbar(data=df_for_display, 
                 x=xvar, y=yvar,
                 xerr=df_for_display[['lower', 'upper']].T, ls='none')
    plt.ylabel(ylab)
    plt.xlabel(xlab)
    plt.title(title)
    plt.show()
```

```python
for k in data_dict.keys(): 
    for v in vars_to_check.keys(): 
        if v in k:
            xlab = 'Ratio for ' + v
            var = vars_to_check[v]
            
    ylab = k.split('_')[1] + ' Feature'
    df_for_display = data_dict[k].reset_index()
    df_for_display = df_for_display.loc[df_for_display.credible == 'yes']
    fig, ax = plt.subplots(figsize=(6, 10))
    display_top_n_feats(df_for_display, 25, 'Feature', var+'mean', xlab, ylab,
                        'Top Features: ' + k)
```
