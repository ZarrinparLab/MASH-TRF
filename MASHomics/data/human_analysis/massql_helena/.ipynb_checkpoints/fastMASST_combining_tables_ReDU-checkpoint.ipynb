```python
# define the folder path
folder_path_nosync = '/Users/helenarusso/Documents/Pesquisa/UCSD Post-doc/Stephany Flores-Ramos/fastMASST_searches'

# get the list of filenames in the folder
filenames_nosync = os.listdir(folder_path_nosync)

# create a text file with the filenames
with open("filenames.txt", "w") as f:
    f.write("\n".join(filenames_nosync))
    
#Create dataframe 
filenames_df_nosync = pd.DataFrame(filenames_nosync, columns=['filename'])
filenames_df_nosync['Scan'] = filenames_df_nosync['filename'].str.split('_').str[3] # make sure you put the correct
#number here on the strip and the proper separator in your case so that Scan column will indeed contain the scan.
```

```python
filenames_df_nosync #make sure that the scan was split into another column.
```

```python
len(filenames_df_nosync['Scan'].unique())
#here is just to compare with how many scans there were in the beginning, and how many ran successfully
```

```python
#let's categorize all results
microbe = filenames_df_nosync[filenames_df_nosync['filename'].str.contains('microbe.tsv')]
food = filenames_df_nosync[filenames_df_nosync['filename'].str.contains('food.tsv')]
plant = filenames_df_nosync[filenames_df_nosync['filename'].str.contains('plant.tsv')]
datasets = filenames_df_nosync[filenames_df_nosync['filename'].str.contains('datasets.tsv')]
matches = filenames_df_nosync[filenames_df_nosync['filename'].str.contains('matches.tsv')]
```

```python
#how many scans that were ran had results in each domain?

print(len(microbe))
print(len(food))
print(len(plant))
print(len(datasets))
print(len(matches))
```

```python
#specify the domains to consider
domains = ['microbe.tsv', 'food.tsv', 'plant.tsv', 'datasets.tsv', 'matches.tsv']

#combine all the files within each domain
output_file = '/Users/helenarusso/Documents/Pesquisa/UCSD Post-doc/Stephany Flores-Ramos/combined_tables/Stephany_'

for domain in domains:
    i = 0
    for file in filenames_nosync: 
        if file.endswith(domain):
            with open(os.path.join(folder_path_nosync, file), 'r') as f:
                with open(output_file + domain, 'a') as out_f:
                    scan = file.split('/')[-1].split('_')[3] 
                    if i == 0:
                        header = next(f).strip()
                        print(header + '\tScan', file=out_f)
                    else:
                        next(f)
                    for line in f:
                        print(line.strip() + '\t' + scan, file=out_f)
                    i += 1
```

# ReDU merging

```python
#import - download database from here https://redu.gnps2.org/
df_redu = pd.read_csv('/Users/helenarusso/Documents/Pesquisa/UCSD Post-doc/Acyl lipids/ReDU_2024_02_17_all_sampleinformation.tsv', sep='\t')
```

```python
#remove 'f.' from filename column, create new column with the dataset and last element of the filename column, and 
#remove the extension of the file (.mzML/.mzML) to match the table we have in the other table.
df_redu['filename'] = df_redu['filename'].str.replace('f.', '')
df_redu['filename_2'] = df_redu['filename'].str.split('/').str[-1]
df_redu['filename_2'] = df_redu['filename_2'].str.replace('.mzML', '')
df_redu['filename_2'] = df_redu['filename_2'].str.replace('.mzXML', '')
df_redu['filepath'] = df_redu['ATTRIBUTE_DatasetAccession'].astype(str).str.cat(df_redu[['filename_2']].astype(str),
                                                                                sep='/')
```

```python
#create 'filepath' column in the fastmasst_matches dataframe

fastmasst_matches['filepath'] = fastmasst_matches['USI'].str.split(':').str[1] + '/' + fastmasst_matches['USI'].str.split(':').str[2]

#merge both
fastmasst_matches_redu = pd.merge(fastmasst_matches, df_redu, on='filepath', how='left')
```

```python
fastmasst_matches_redu.head()
```

```python
#filter to have only human-related datasets
matches_homo_sapiens = fastmasst_matches_redu.loc[fastmasst_matches_redu['NCBITaxonomy'] == '9606|Homo sapiens']
```
